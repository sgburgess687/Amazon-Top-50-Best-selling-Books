# Explore Name

[Back to top](#Top)

```python
df.Name.value_counts()
```

None of the books are included in all of the Top 50 charts across the years.

The closest is the *APA Publication Manual*, which appears in 10 of the 11 years.

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(data=df, x=df.Name.value_counts(), palette="crest_r")

for p in ax.patches:
    ax.annotate(format(p.get_height(), 'd'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

ax.set(xlabel="Appearances", ylabel="Count")
ax.set_title("Times a Title Shows up on Top 50 Charts")
plt.show()
```

A vast majority of the books are unique to a single year's Top 50.

```python
df.loc[df['Name'] == 'Publication Manual of the American Psychological Association, 6th Edition']
```

It looks like the the *APA Publication Manual* made the Top 50 in every year except 2019.

# Explore Author

[Back to top](#Top)

```python
df.Author.value_counts()
```

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(data=df, x=df.Author.value_counts(), palette="crest_r")

for p in ax.patches:
    ax.annotate(format(p.get_height(), 'd'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

ax.set(xlabel="Appearances", ylabel="Count")
ax.set_title("Total Appearances in the Top 50 Charts")
plt.show()
```

Clearly, most of the authors appear only once in the top 50 charts.

### 50 Authors that Appeared the Most

```python
df.Author.value_counts()[:50]
```

```python
# which books did Dr. Seuss have in the top 50?
df.loc[df['Author'] == 'Dr. Seuss']
```

Dr. Seuss only had two books across the top 50 bestsellers,
but "Oh, the Places You'll Go!" reached top 50 a total of 8 times.

### Authors with Repeat Top 50 Appearances

Let's see who made the top 50 more than once:

```python
df_repeat = pd.DataFrame(
            [x, y] for x, y in df.Author.value_counts().iteritems() if (y > 1)
)
```

```python
df_repeat.columns = ['Author','Top 50 Appearances']
```

```python
df_repeat
```

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(data=df_repeat, x="Top 50 Appearances", palette="crest_r")

for p in ax.patches:
    ax.annotate(format(p.get_height(), 'd'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

ax.set(xlabel="Times in Top 50", ylabel="Count")
ax.set_title("Repeat Top 50 Appearances")
plt.show()
```

Of those who made the Top 50 more than once, a majority only reached it twice.

```python
cols = ['Author', 'Count']
dat = pd.DataFrame(columns = cols)
for name, count in df_2019.Author.value_counts().items():
    if count > 1:
        dat = dat.append({'Author': str(name), 'Count':count},ignore_index=True)
```

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(data=df_repeat, x="Top 50 Appearances", palette="crest_r")

for p in ax.patches:
    ax.annotate(format(p.get_height(), 'd'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

ax.set(xlabel="Times in Top 50", ylabel="Count")
ax.set_title("Repeat Top 50 Appearances")
plt.show()
```

```python
plt.figure(figsize=(10,6))


ax = dat.plot(kind='bar', x='Author', rot=0, colormap='jet')

for p in ax.patches:
    ax.annotate(format(p.get_height(), 'd'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 5), 
                   textcoords = 'offset points')

ax.set(xlabel="Author", ylabel="Count")
ax.set_title("2019 Repeat Appearances")
plt.show()
```
